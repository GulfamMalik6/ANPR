```python
# Import the necessary libraries
import easyocr  # Import the EasyOCR library for text recognition
import cv2      # Import OpenCV for image processing
import numpy as np  # Import numpy for numerical operations
from matplotlib import pyplot as plt  # Import Matplotlib for image visualization
```

```python
# Create an EasyOCR Reader configured for English text recognition
reader = easyocr.Reader(['en'])

# Define the path to the image we want to process
IMAGE_PATH = "/text.png"

# Use the Reader object to perform text recognition on the image
result = reader.readtext(IMAGE_PATH)
```

```python
# Extract the coordinates of the top-left and bottom-right corners of the bounding box
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])

# Extract the detected text from the result
text = result[0][1]
```

```python
img = cv2.imread(IMAGE_PATH)

# Extract detected texts and their coordinates
detected_texts = [(detection[1], tuple(detection[0][0]), tuple(detection[0][2])) for detection in result]

# Display the image
plt.imshow(img)
plt.show()

# Print detected texts
for i, (text, _, _) in enumerate(detected_texts):
    print(f"Detected Text {i + 1}: {text}")
```
